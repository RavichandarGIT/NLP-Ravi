```python
# import libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
```

```python
# download required dataset
nltk.download('punkt_tab')
```

```python
# sample text
text = " natrual language processing enables machines to understand human languages undeestanding "
```

```python
# tokenization
tokens = word_tokenize(text)
```

```python
# frequency Distribution
freq_dist = FreqDist(tokens)
```

```python
# display word frequencies
print("Word Frequencies:", freq_dist)
print("Most common Word:", freq_dist.most_common(5))
print("Vocabulary Size:", len(freq_dist))

```
